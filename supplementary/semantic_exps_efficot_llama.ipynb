```python
from tqdm import tqdm
import sys
sys.path.append("..")
from inference.inference import get_formatted_prompt
from models.contemp_generator import ContemplationGenerator
from transformers import AutoTokenizer, AutoModelForCausalLM

def get_contemp_tokens(variations, dataset, config_name, device, eval_num_contemp_tokens=1):
    if config_name == "small":
        teacher_model_name = "meta-llama/Llama-2-7b-chat-hf"
    elif config_name == 'mistral':
        teacher_model_name = "mistralai/Mistral-7B-Instruct-v0.2"

    model_save_path = f"/data/nee7ne/effi_cot/saved_models/effi_cot/vanilla/small_old/{dataset}"
    contemp_generator = ContemplationGenerator.from_pretrained(model_save_path+"/contemp_generator").to(device)

    teacher_tokenizer = AutoTokenizer.from_pretrained(teacher_model_name)
    teacher_tokenizer.pad_token = teacher_tokenizer.eos_token
    teacher_model = AutoModelForCausalLM.from_pretrained(teacher_model_name)
    teacher_model = teacher_model.to(device)
    teacher_model.eval()

    contemp_tokens = []
    for variation in tqdm(variations):
        query_prompt, answer_prompt = get_formatted_prompt(variation)
        # Find the position where we inserted the contemplation tokens
        query_inputs = contemp_generator.tokenizer(
            query_prompt,
            return_tensors="pt",
            padding=False,
            truncation=False,
            max_length=512
        ).to(device)
        prefix_length = query_inputs.input_ids.size(1) - 1

        answer_inputs = contemp_generator.tokenizer(
            answer_prompt,
            return_tensors="pt",
            padding=False,
            truncation=True,
            max_length=512,
            add_special_tokens=False
        ).to(device)

        # CHANGED: Extract contemplation states from the correct position (no longer the last tokens)
        query_inputs = torch.cat([
            query_inputs['input_ids'],
            torch.tensor([[contemp_generator.tokenizer.eos_token_id * eval_num_contemp_tokens]]).to(device),
            answer_inputs['input_ids']
        ], dim=1)

        # Get contemplation states from the correct position
        contemp_states = contemp_generator(
            query_inputs,
            attention_mask=torch.ones_like(query_inputs)
        )[:, prefix_length:prefix_length+eval_num_contemp_tokens, :]
        contemp_tokens.append(contemp_states.detach().cpu())
    # import pdb
    # pdb.set_trace()
    del contemp_generator, contemp_states, query_inputs, answer_inputs
    torch.cuda.empty_cache()
    return contemp_tokens
```

```python
for dataset in datasets:
    for sample in datasets[dataset]:
        sample["tokens"] = get_contemp_tokens(sample['queries'], dataset, "small", "cuda:3")
```
